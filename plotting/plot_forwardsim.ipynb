```python
#####################################
########### TO SPECIFY ##############
#####################################

save_dir = '../run_0'
which_agent = 4
time_lim = 100

#which agg iters to plot
start_aggregation_iter = 0
stop_aggregation_iter = 1
agg_increment = 1

#####################################
#####################################

#imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from itertools import cycle

#init vars
lines = ["--","-.",":"]
linecycler = cycle(lines)
linestyle_truth = 'k-'
linewidth_truth = 3
linewidth_pred = 2
fig_size = (10, 50)
agg_iter = start_aggregation_iter

while(agg_iter <= stop_aggregation_iter):
    
    ### Read in forwardsim data 
    states_true = np.load(save_dir + '/saved_forwardsim/forwardsim_states_true_'+str(agg_iter)+'.npy')
    states_pred = np.load(save_dir + '/saved_forwardsim/forwardsim_states_pred_'+str(agg_iter)+'.npy')
    
    ### Select linetype for plotting
    linestyle= next(linecycler)
    
    

    ### Visualizing forward simulation of ant
    ### state (41) = 
        # 15 joint pos, 14 joint vel
        # 3 com pos
        # 9 entries of rotation matrix
        # 3 com vel

    if (which_agent==1):
        
        if(agg_iter == start_aggregation_iter):
            #f, ax = plt.subplots(41, 1, figsize=fig_size) #if you want to plot everything
            f, ax = plt.subplots(15, 1, figsize=fig_size) #not plotting everything

        ax[0].set_title('x com')
        ax[0].plot(states_true[0:time_lim, 29], linestyle_truth, linewidth = linewidth_truth)
        ax[0].plot(states_pred[0:time_lim, 29], linestyle, linewidth=linewidth_pred, label='Agg iter '+str(agg_iter))
        ax[0].legend()

        ax[1].set_title('y com')
        ax[1].plot(states_true[0:time_lim, 30], linestyle_truth, linewidth = linewidth_truth)
        ax[1].plot(states_pred[0:time_lim, 30], linestyle, linewidth=linewidth_pred)
        
        ax[2].set_title('z com')
        ax[2].plot(states_true[0:time_lim, 31], linestyle_truth, linewidth = linewidth_truth)
        ax[2].plot(states_pred[0:time_lim, 31], linestyle, linewidth=linewidth_pred)
        
        ax[3].set_title('x com vel')
        ax[3].plot(states_true[0:time_lim, 38], linestyle_truth, linewidth = linewidth_truth)
        ax[3].plot(states_pred[0:time_lim, 38], linestyle, linewidth=linewidth_pred, label='Agg iter '+str(agg_iter))

        ax[4].set_title('y com vel')
        ax[4].plot(states_true[0:time_lim, 39], linestyle_truth, linewidth = linewidth_truth)
        ax[4].plot(states_pred[0:time_lim, 39], linestyle, linewidth=linewidth_pred)
        
        ax[5].set_title('z com vel')
        ax[5].plot(states_true[0:time_lim, 40], linestyle_truth, linewidth = linewidth_truth)
        ax[5].plot(states_pred[0:time_lim, 40], linestyle, linewidth=linewidth_pred)
        
        ax[6].set_title('roll')
        roll_angles_true = np.arctan2(states_true[0:time_lim, 33], states_true[0:time_lim, 32])
        roll_angles_pred = np.arctan2(states_pred[0:time_lim, 33], states_pred[0:time_lim, 32])
        ax[6].plot(roll_angles_true[0:time_lim], linestyle_truth, linewidth = linewidth_truth)
        ax[6].plot(roll_angles_pred[0:time_lim], linestyle, linewidth=linewidth_pred)
        
        ax[7].set_title('pitch')
        pitch_angles_true = np.arctan2(states_true[0:time_lim, 35], states_true[0:time_lim, 34])
        pitch_angles_pred = np.arctan2(states_pred[0:time_lim, 35], states_pred[0:time_lim, 34])
        ax[7].plot(pitch_angles_true[0:time_lim], linestyle_truth, linewidth = linewidth_truth)
        ax[7].plot(pitch_angles_pred[0:time_lim], linestyle, linewidth=linewidth_pred)
        
        ax[8].set_title('yaw')
        yaw_angles_true = np.arctan2(states_true[0:time_lim, 37], states_true[0:time_lim, 36])
        yaw_angles_pred = np.arctan2(states_pred[0:time_lim, 37], states_pred[0:time_lim, 36])
        ax[8].plot(yaw_angles_true[0:time_lim], linestyle_truth, linewidth = linewidth_truth)
        ax[8].plot(yaw_angles_pred[0:time_lim], linestyle, linewidth=linewidth_pred)
        
        ax[9].set_title('j1 pos')
        ax[9].plot(states_true[0:time_lim, 0], linestyle_truth, linewidth = linewidth_truth)
        ax[9].plot(states_pred[0:time_lim, 0], linestyle, linewidth=linewidth_pred, label='Agg iter '+str(agg_iter))

        ax[10].set_title('j2 pos')
        ax[10].plot(states_true[0:time_lim, 1], linestyle_truth, linewidth = linewidth_truth)
        ax[10].plot(states_pred[0:time_lim, 1], linestyle, linewidth=linewidth_pred)
        
        ax[11].set_title('j3 pos')
        ax[11].plot(states_true[0:time_lim, 2], linestyle_truth, linewidth = linewidth_truth)
        ax[11].plot(states_pred[0:time_lim, 2], linestyle, linewidth=linewidth_pred)
        
        ax[12].set_title('j1 vel')
        ax[12].plot(states_true[0:time_lim, 15], linestyle_truth, linewidth = linewidth_truth)
        ax[12].plot(states_pred[0:time_lim, 15], linestyle, linewidth=linewidth_pred, label='Agg iter '+str(agg_iter))

        ax[13].set_title('j2 vel')
        ax[13].plot(states_true[0:time_lim, 16], linestyle_truth, linewidth = linewidth_truth)
        ax[13].plot(states_pred[0:time_lim, 16], linestyle, linewidth=linewidth_pred)
        
        ax[14].set_title('j3 vel')
        ax[14].plot(states_true[0:time_lim, 17], linestyle_truth, linewidth = linewidth_truth)
        ax[14].plot(states_pred[0:time_lim, 17], linestyle, linewidth=linewidth_pred)
        


    ### Visualizing forward simulation of swimmer
    ### state (16): 
        # slider x, slider y, heading, 2 hinge joint pos, 
        # 2 slider x/y vel, heading vel, 2 hinge joint vel, 
        # 3 com pos, 3 com vel

    if (which_agent==2):

        if(agg_iter == start_aggregation_iter):
            f, ax = plt.subplots(16, 1, figsize=fig_size)

        ax[0].set_title('x com')
        ax[0].plot(states_true[0:time_lim, 10], linestyle_truth, linewidth = linewidth_truth)
        ax[0].plot(states_pred[0:time_lim, 10], linestyle, linewidth=linewidth_pred, label='Agg iter '+str(agg_iter))
        ax[0].legend()

        ax[1].set_title('y com')
        ax[1].plot(states_true[0:time_lim, 11], linestyle_truth, linewidth = linewidth_truth)
        ax[1].plot(states_pred[0:time_lim, 11], linestyle, linewidth=linewidth_pred)

        ax[2].set_title('z com')
        ax[2].plot(states_true[0:time_lim, 12], linestyle_truth, linewidth = linewidth_truth)
        ax[2].plot(states_pred[0:time_lim, 12], linestyle, linewidth=linewidth_pred)

        ax[3].set_title('x slider pos')
        ax[3].plot(states_true[0:time_lim, 0], linestyle_truth, linewidth = linewidth_truth)
        ax[3].plot(states_pred[0:time_lim, 0], linestyle, linewidth=linewidth_pred)

        ax[4].set_title('y slider pos')
        ax[4].plot(states_true[0:time_lim, 1], linestyle_truth, linewidth = linewidth_truth)
        ax[4].plot(states_pred[0:time_lim, 1], linestyle, linewidth=linewidth_pred)

        ax[5].set_title('heading')
        ax[5].plot(states_true[0:time_lim, 2], linestyle_truth, linewidth = linewidth_truth)
        ax[5].plot(states_pred[0:time_lim, 2], linestyle, linewidth=linewidth_pred)

        ax[6].set_title('j1 pos')
        ax[6].plot(states_true[0:time_lim, 3], linestyle_truth, linewidth = linewidth_truth)
        ax[6].plot(states_pred[0:time_lim, 3], linestyle, linewidth=linewidth_pred)

        ax[7].set_title('j2 pos')
        ax[7].plot(states_true[0:time_lim, 4], linestyle_truth, linewidth = linewidth_truth)
        ax[7].plot(states_pred[0:time_lim, 4], linestyle, linewidth=linewidth_pred)

        ax[8].set_title('x slider vel')
        ax[8].plot(states_true[0:time_lim, 5], linestyle_truth, linewidth = linewidth_truth)
        ax[8].plot(states_pred[0:time_lim, 5], linestyle, linewidth=linewidth_pred)

        ax[9].set_title('y slider vel')
        ax[9].plot(states_true[0:time_lim, 6], linestyle_truth, linewidth = linewidth_truth)
        ax[9].plot(states_pred[0:time_lim, 6], linestyle, linewidth=linewidth_pred)

        ax[10].set_title('heading vel')
        ax[10].plot(states_true[0:time_lim, 7], linestyle_truth, linewidth = linewidth_truth)
        ax[10].plot(states_pred[0:time_lim, 7], linestyle, linewidth=linewidth_pred)

        ax[11].set_title('j1 vel')
        ax[11].plot(states_true[0:time_lim, 8], linestyle_truth, linewidth = linewidth_truth)
        ax[11].plot(states_pred[0:time_lim, 8], linestyle, linewidth=linewidth_pred)

        ax[12].set_title('j2 vel')
        ax[12].plot(states_true[0:time_lim, 9], linestyle_truth, linewidth = linewidth_truth)
        ax[12].plot(states_pred[0:time_lim, 9], linestyle, linewidth=linewidth_pred)

        ax[13].set_title('x com vel')
        ax[13].plot(states_true[0:time_lim, 13], linestyle_truth, linewidth = linewidth_truth)
        ax[13].plot(states_pred[0:time_lim, 13], linestyle, linewidth=linewidth_pred)

        ax[14].set_title('y com vel')
        ax[14].plot(states_true[0:time_lim, 14], linestyle_truth, linewidth = linewidth_truth)
        ax[14].plot(states_pred[0:time_lim, 14], linestyle, linewidth=linewidth_pred)

        ax[15].set_title('z com vel')
        ax[15].plot(states_true[0:time_lim, 15], linestyle_truth, linewidth = linewidth_truth)
        ax[15].plot(states_pred[0:time_lim, 15], linestyle, linewidth=linewidth_pred)
        
    ### Visualizing forward simulation of cheetah
    ### state (24): 
        # rootx (forward/backward), rootz (up/down), rooty (angle)
        # bthigh, bshin, bfoot, fthigh, fshin, ffoot
        # rootx vel, rootz vel, rooty vel
        # bthigh vel, bshin vel, bfoot vel, fthigh vel, fshin vel, ffoot vel
        # 3 com pos, 3 com vel

    if (which_agent==4):

        if(agg_iter == start_aggregation_iter):
            #f, ax = plt.subplots(24, 1, figsize=fig_size) #plotting everything
            f, ax = plt.subplots(18, 1, figsize=fig_size) #not plotting everything

        ax[0].set_title('x com')
        ax[0].plot(states_true[0:time_lim, 18], linestyle_truth, linewidth = linewidth_truth)
        ax[0].plot(states_pred[0:time_lim, 18], linestyle, linewidth=linewidth_pred, label='Agg iter '+str(agg_iter))
        ax[0].legend()

        ax[1].set_title('y com')
        ax[1].plot(states_true[0:time_lim, 19], linestyle_truth, linewidth = linewidth_truth)
        ax[1].plot(states_pred[0:time_lim, 19], linestyle, linewidth=linewidth_pred)

        ax[2].set_title('z com')
        ax[2].plot(states_true[0:time_lim, 20], linestyle_truth, linewidth = linewidth_truth)
        ax[2].plot(states_pred[0:time_lim, 20], linestyle, linewidth=linewidth_pred)
        
        ax[3].set_title('x com vel')
        ax[3].plot(states_true[0:time_lim, 21], linestyle_truth, linewidth = linewidth_truth)
        ax[3].plot(states_pred[0:time_lim, 21], linestyle, linewidth=linewidth_pred)

        ax[4].set_title('y com vel')
        ax[4].plot(states_true[0:time_lim, 22], linestyle_truth, linewidth = linewidth_truth)
        ax[4].plot(states_pred[0:time_lim, 22], linestyle, linewidth=linewidth_pred)

        ax[5].set_title('z com vel')
        ax[5].plot(states_true[0:time_lim, 23], linestyle_truth, linewidth = linewidth_truth)
        ax[5].plot(states_pred[0:time_lim, 23], linestyle, linewidth=linewidth_pred)
        
        ax[6].set_title('bthigh')
        ax[6].plot(states_true[0:time_lim, 3], linestyle_truth, linewidth = linewidth_truth)
        ax[6].plot(states_pred[0:time_lim, 3], linestyle, linewidth=linewidth_pred)

        ax[7].set_title('bshin')
        ax[7].plot(states_true[0:time_lim, 4], linestyle_truth, linewidth = linewidth_truth)
        ax[7].plot(states_pred[0:time_lim, 4], linestyle, linewidth=linewidth_pred)

        ax[8].set_title('bfoot')
        ax[8].plot(states_true[0:time_lim, 5], linestyle_truth, linewidth = linewidth_truth)
        ax[8].plot(states_pred[0:time_lim, 5], linestyle, linewidth=linewidth_pred)
        
        ax[9].set_title('fthigh')
        ax[9].plot(states_true[0:time_lim, 6], linestyle_truth, linewidth = linewidth_truth)
        ax[9].plot(states_pred[0:time_lim, 6], linestyle, linewidth=linewidth_pred)

        ax[10].set_title('fshin')
        ax[10].plot(states_true[0:time_lim, 7], linestyle_truth, linewidth = linewidth_truth)
        ax[10].plot(states_pred[0:time_lim, 7], linestyle, linewidth=linewidth_pred)

        ax[11].set_title('ffoot')
        ax[11].plot(states_true[0:time_lim, 8], linestyle_truth, linewidth = linewidth_truth)
        ax[11].plot(states_pred[0:time_lim, 8], linestyle, linewidth=linewidth_pred)
        
        ax[12].set_title('bthigh vel')
        ax[12].plot(states_true[0:time_lim, 12], linestyle_truth, linewidth = linewidth_truth)
        ax[12].plot(states_pred[0:time_lim, 12], linestyle, linewidth=linewidth_pred)

        ax[13].set_title('bshin vel')
        ax[13].plot(states_true[0:time_lim, 13], linestyle_truth, linewidth = linewidth_truth)
        ax[13].plot(states_pred[0:time_lim, 13], linestyle, linewidth=linewidth_pred)

        ax[14].set_title('bfootl vel')
        ax[14].plot(states_true[0:time_lim, 14], linestyle_truth, linewidth = linewidth_truth)
        ax[14].plot(states_pred[0:time_lim, 14], linestyle, linewidth=linewidth_pred)
        
        ax[15].set_title('fthigh vel')
        ax[15].plot(states_true[0:time_lim, 15], linestyle_truth, linewidth = linewidth_truth)
        ax[15].plot(states_pred[0:time_lim, 15], linestyle, linewidth=linewidth_pred)

        ax[16].set_title('fshin vel')
        ax[16].plot(states_true[0:time_lim, 16], linestyle_truth, linewidth = linewidth_truth)
        ax[16].plot(states_pred[0:time_lim, 16], linestyle, linewidth=linewidth_pred)

        ax[17].set_title('ffoot vel')
        ax[17].plot(states_true[0:time_lim, 17], linestyle_truth, linewidth = linewidth_truth)
        ax[17].plot(states_pred[0:time_lim, 17], linestyle, linewidth=linewidth_pred)
        
        
    
    agg_iter+=agg_increment
```
